
# 02. Tipos de datos
Los arrays de NumPy solo pueden contener un tipo de dato, ya que esto es lo que le confiere las ventajas de la optimización de memoria.

### Importando librerias

```python
import numpy as np
```

```python
arr = np.array([1,2,3,4])
arr.dtype # Si queremos usar otro tipo de dato lo podemos definir en la declaración del array.
```

```python
arr = np.array([1,2,3,4], dtype='float64')
arr.dtype # Ahora vemos que los valores están con punto decimal.
```

```python
arr # Si ya se tiene el array definido se usa el método .astype() para convertir el tipo de dato.
```

```python
arr = np.array([1, 2, 3, 4])
arr = arr.astype(np.float64)
arr # También se puede cambiar a tipo booleano recordando que los números diferentes de 0 se convierten en True.
```

### Cambiando tipo de datos

```python
arr = np.array([0,1,2,3,4])
arr = arr.astype(np.bool_)
arr # También podemos convertir los datos en tipo string.
```

```python
arr = np.array([0,1,2,3,4])
arr = arr.astype(np.string_)
arr # De igual forma se puede pasar de string a numero.
```

```python
arr = np.array(['hola','1','2','3','4'])
arr = arr.astype(np.int8)
arr # Si un elemento no es de tipo número el método falla.'''
```
