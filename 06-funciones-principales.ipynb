### Percentile

```python
np.median(arr) # La mediana de un arreglo. Que significa el percentil 50, donde se encuentra la mitad del array
```

### Mean

### Concatenate

```python
a = np.array([[1,2],[3,4]])
b = np.array([5,6])
```

```python
np.concatenate((a,b),axis=0) # Se concatena una tupla por el eje 0
# El error es debido a que "a" tiene 2 dimensiones mientras que "b" tiene 1.
```

```python
a.ndim
b.ndim
```

```python
b = np.expand_dims(b,axis=0)
```

```python
np.concatenate((a,b),axis=0)
```

```python
b
```

```python
np.concatenate((a,b),axis=1)
# Como "b" es una fila y no una columna, no se puede concatenar a menos que se aplique la Tanspuesta (T).
```

```python
np.concatenate((a,b.T),axis=1) # Se aplica la Transpuesta de una matriz (T)
```
