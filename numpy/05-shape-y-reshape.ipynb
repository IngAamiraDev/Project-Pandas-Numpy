# 05. 🟦🔹Shape y Reshape
- Shape   => Me indica la forma del arreglo
- Reshape => Transforma el arreglo mientras se mantengan los elementos.

### Importar librerías

```python
import numpy as np 
```

```python
arr = np.random.randint(1,10,(3,2)) # Array de 1 a 10 con una estructura de 3 x 2
```

### Shape

```python
arr.shape # Indica la cantidad de filas y columnas. Muy útil para saber que tan grande es el arreglo que se está trabajando.
```

```python
arr
```

### Reshape

```python
arr.reshape(1,6) # Comando para cambiar la forma establecida en un array
```

```python
arr.reshape(2,3)
```

```python
np.reshape(arr,(1,6))
```

```python
arr
```

```python
np.reshape(arr,(2,3),'C') # Se puede hacer un reshape como lo haría en el lengaje de C.
```

```python
np.reshape(arr,(2,3),'F') # Se puede hacer un reshape como lo haría en el lengaje de Fortran.
```

```python
np.reshape(arr,(2,3),'A') # Además existe la opción de hacer reshape según como esté optimizado el guardado de datos en memoría del sistema. En este caso es como en C.
```
