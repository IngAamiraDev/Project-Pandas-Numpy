# 05. Shape y Reshape
- Shape   => Me indica la forma del arreglo
- Reshape => Transforma el arreglo mientras se mantengan los elementos.

### Importar librer铆as

```python
import numpy as np 
```

```python
arr = np.random.randint(1,10,(3,2)) # Array de 1 a 10 con una estructura de 3 x 2
```

### Shape

```python
arr.shape # Indica la cantidad de filas y columnas. Muy 煤til para saber que tan grande es el arreglo que se est谩 trabajando.
```

### Reshape

```python
arr.reshape(1,6) # Comando para cambiar la forma establecida en un array
```

```python
arr.reshape(2,3)
```

```python
np.reshape(arr,(1,6))
```
