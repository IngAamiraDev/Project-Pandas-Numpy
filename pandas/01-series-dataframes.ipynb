### Importar librería

```python
import pandas as pd 
```

### Pandas Series

Es muy parecido a un array de una dimensión (o vector) de NumPy.
- Arreglo unidimensional indexado
- Búsqueda por índice
- Slicing
- Operaciones aritméticas
- Distintos tipos de datos

### Crear Series

```python
psg_players = pd.Series(['Navas','Mbappe','Neymar','Messi'],
          index=[1,7,10,30] # Se define los índices para acceder a la data.
          )
```

```python
psg_players
```

```python
psg_players = pd.Series(['Navas','Mbappe','Neymar','Messi'])
```

```python
dict = {1:'Navas',7:'Mbappe',10:'Neymar',30:'Messi'} # Se define como un diccionario.
pd.Series(dict) # Entiende cuál es la estructura que se va a menejar.
```

```python
psg_players[0:3] # Se hace un Slicing
```

### Pandas DataFrame

Muy parecido a las estructuras matriciales trabajadas con NumPy.

- Estructura principal
- Arreglo de dos dimensiones
- Búsqueda por índice (columnas o filas)
- Slicing
- Operaciones aritméticas
- Distintos tipos de datos
- Tamaño variable

### Crear DataFrame

```python
dict = { # Estructura de diccionario matricial de 3 dimensiones
    'Jugador':['Navas','Mbappe','Neymar','Messi'],
    'Altura':[183.0, 170.0, 170.0, 165.0],
    'Goles':[2,200,200,200]
 }
```

```python
pd.DataFrame(dict,index=[1,7,10,30]) # Dataframe con index personalizado
```

```python
pd.DataFrame(dict) # Dataframe con index default
```

```python
df_Players = pd.DataFrame(dict,index=[1,7,10,30])
```
