# 06. Funciones principales de Pandas
- Con la función info puedo obtener un resumen de como esta compuesto mi dataset, es una buena idea iniciar con está función al analisar nuestros datos
- Con la función describe puedo saber cierta información estadística, solamente de nuestros datos numéricos. Esto es muy usado en la ciencia de datos
- La función head me retorna los primeros registros que yo especifique y la función tail me retorna los ultimos registros que yo especifique
- Con la función memory_usage junto con el parametro deep=True, puedo saber cuanta memoria estoy usando en mis DataFrame
- Con la función value_counts y especificando mi objeto, puedo contar los valores que existen de ese objeto
- Con la función drop_duplicates puedo eliminar los duplicados que se encuentran en mi dataset
- Con la función sort_values puedo organizar mis datos ascendentemente o descendentemente según la columna que yo especifique
- Existen muchas funciones que es importante practicar

## Importar librería

```python
import pandas as pd
```

## Cargar DataFrame

```python
df_books = pd.read_csv('./bd/bestsellers-with-categories.csv',sep=',',header=0)
df_books.head(2) # Mostrar las primeras dos lineas de registro
```

## info()

```python
df_books.info() # Mostrar los diferentes datos que contiene el dataFrame.
```

## describe()

```python
df_books.describe() # Obtener diferentes datos estadisticos de las columnas numericas.
```

## head() y tail()

```python
df_books.tail(5) # Mostrar los últimos 5 registros del dataFrame
```

```python
df_books.head(5)
```

## memory_usage()

```python
df_books.memory_usage(deep=True) # Obtener el uso de la memoria de cada columna
```

## value_counts()

```python
df_books['Author'].value_counts() # Cuenta cuantos valores tengo por cada uno de los autores
```

## drop_duplicates()

```python
df_books.iloc[0] # Obtener la informacion de un registro especifico
```

```python
df_books = df_books.append(df_books.iloc[0]) # Se usa para incrementar el dataframe
```

```python
df_books.drop_duplicates() # Eliminar registros duplicados
```

```python
df_books.drop_duplicates(keep='last') # Elimina el primer registro duplicado del dataFrame (orden ascendente)
```

## sort_values()

```python
df_books.sort_values('Year') # Ordena los valores de menor a mayor segun el año (orden ascendente)
```

```python
df_books.sort_values('Year',ascending=False) # Ordenar los registros segun valores de la columna (orden desendiente)
```
